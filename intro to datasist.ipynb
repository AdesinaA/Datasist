# Working with Datasist structdata

You ever imagine, What is Datasist?. Well, I will go by saying: Datasist is a python library that makes easy data analysis, visualization,cleaning andpreparation for data scientist during prototyping.

Note: We will be using Jupyter notebook and will be working with Data science Nigeria 2020 hackhaton dataset. After getting all this ready. Let's dive into action.

Finally, open your jupyter notebook, import your libraries and dataset as shown below:

```python
import pandas as pd
import numpy as np
import datasist as ds      

train_df = pd.read_csv(r'C:\Users\Grade A\Desktop\MY HUB\DATASET\DSN\train.csv')
test_df =pd.read_csv(r'C:\Users\Grade A\Desktop\MY HUB\DATASET\DSN\test.csv')
```

The structdata moddule has many functions for working with structured data in the Pandas DataFrame format. So, you can easily manipulate and analyze DataFrames. Let's dive into in the functions available.

1. Describe: We all that Pandas also has describe function in Pandas. Let's take a look at Datasist describe method.

```python
ds.structdata.describe(train_df)
```

2. check_train_test_set: To use this function, (train_df and test_df) function muxt be passed. A common index(Apllicant_ID) and any feature availabe in both dataset.

```python
ds.structdata.check_train_test_set(train_df, test_df, index='Applicant_ID', col='form_field47')
```

3. display_missing values: You can check for the missing values in the dataset. Let's check it out.

```python
ds.structdata.display_missing(train_df)
```

4. get_cats_feats and get_num_feats : This functions are used to retreive categorical and numerical features and they give their output as a list.

```python
cat_feats = ds.structdata.get_cat_feats(train_df)
cat_feats
```

```python
num_feats = ds.structdata.get_num_feats(train_df)
num_feats
```

5. get_unique_counts : In a categorical features, you can the function to get the unique class. Let's take a look

```python
ds.structdata.get_unique_counts(train_df)
```

6. join_train_test : You can use the function (join_train_and_test) function to concatenate both the train and test set.

```python
datas, btrain, btest = ds.structdata.join_train_and_test(train_df, test_df)
print("New size of combined data {}".format(datas.shape))
print("Old size of train data: {}".format(btrain))
print("Old size of test data: {}".format(btest))

#later splitting after transformations
train = datas[:btrain]
test = datas[btrain:]
```

# Feature engineering with Datasist

Feature engineering is the process of using data’s domain knowledge to create features that make machine learning algorithms work. It’s the act of extracting important features from raw data and transforming them into formats that are suitable for machine learning.

Now, let's explore some of the feature_engineering module of datasist.

1. drop_missing: The functions drops features with missing values. Let's take a look below:

```python
#let's display the mssing values first
ds.structdata.display_missing(train_df)
```

```python
#Now, let's drop form_field41.
new_train_df = ds.feature_engineering.drop_missing(train_df, percent=7.0)
ds.structdata.display_missing(new_train_df)
```

2. drop redundant : This functions removes features with no variance, Let's create a new dataset. Take a look:

```python
df = pd.DataFrame({'a': [2,2,2,2,2,2,2],
                  'b': [8,9,10,11,12,13,14]})

df
```

Now, check the dataset. You see that a is redundant which means that it has the same class all through. We will drop the column by using this function. Check below:

```python
df = ds.feature_engineering.drop_redundant(df)
df
```

Ooops, cool. You see that?

3. convert_dtypes: This function takes DataFrame automatically features that are not represented on their right types. Let's have a look:

```python
data = {'Name':['Ade', 'mike', 'wood'],
        'Age':['14', '20', '15']} 
       
df = pd.DataFrame(data)
df
```

Now, let's have a look at the Data types:

```python
df.dtypes
```

Note: The features of Age is supposed to be integer. By using convert_dtype function. It is automatically fixed. Let's have a look:

```python
df = ds.feature_engineering.convert_dtype(df)
df.dtypes
```

4. fill_missing_cats : This function automatically fills in the missing values. Let's check:

```python
#let's look at the mssing vales first
ds.structdata.display_missing(train_df)
```

```python
df = ds.feature_engineering.fill_missing_cats(train_df)
ds.structdata.display_missing(df)
```

5. fill_missing_values : This functions works on numerical features and you can specify a fill strategy (mea, median and mode). Let's check:

```python
df = ds.feature_engineering.fill_missing_num(train_df)
ds.structdata.display_missing(df)
```

# Visualization for Categorical features

Visualization for the categorical features include: violinplot, countplots,boxplot etc. Now, let's start getting our hands on them one after the other.

1. countplot: This makes a barplot of all the categorical feature to show their clas count. Let's check

```python
ds.visualizations.countplot(train_df)
```

```python
ds.visualizations.countplot(train_df, separate_by='default_status')
```

2. boxplot : This makes a box plot of all numerical features against a categorical target column. Have a look:

```python
ds.visualizations.boxplot(train_df, target='default_status')
```

3. catbox: This is used to make a side by side bar plot of all the categorical features in a dataset against a categorical target. Let's check

```python
ds.visualizations.catbox(train_df, target='default_status')
```
